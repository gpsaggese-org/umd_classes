```python
# Import dependencies and dynamically load API class
import importlib.util
import sys
import pandas as pd
import great_expectations as gx
from bitcoin_utils import summarize_validation_result
```

```python
# Load bitcoin.example.py as module
spec = importlib.util.spec_from_file_location("bitcoin_example", "./bitcoin.example.py")
example_module = importlib.util.module_from_spec(spec)
sys.modules["bitcoin_example"] = example_module
spec.loader.exec_module(example_module)
```

```python
# Create an instance of BitcoinMonitor
monitor = example_module.BitcoinMonitor(interval_seconds=3, run_count=5)
```

```python
# Run the data collection and validation loop
monitor.run_loop(verbose=False)
```

```python
# Load data from log file
log_file = "bitcoin_price_log.csv"
df_logged = pd.read_csv(log_file)

# Ensure correct data types for numeric columns
float_cols = [
    "price_usd", "market_cap", "total_volume", "market_cap_rank",
    "circulating_supply", "developer_score", "community_score", "ath", "atl"
]
df_logged[float_cols] = df_logged[float_cols].astype(float)

# Re-validate the entire dataset using Great Expectations
validation_result = monitor.api.validate(df_logged)

# Print validation summary
summarize_validation_result(validation_result)
```

```python
# Build Great Expectations data docs
context = gx.get_context()
context.build_data_docs()
# Display your custom local path to the report
print("Data Docs saved to folder: TutorTask127_Spring2025_Monitoring_Bitcoin_Prices_Using_Great_Expectations\\gx\\uncommitted\\data_docs\\local_site\\index.html")
```
